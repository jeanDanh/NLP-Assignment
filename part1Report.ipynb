Task 1.2.c) Hàm `tacgiaLink()` trích tên tác giả, đường link đến tiểu sử, ngày tháng năm sinh, các câu nói nổi tiếng.

```python

def unduplicateListAuthor(quotesSoup):
    # quotesSoup is the global variable quotes, see pieces of code above to know what it is
    unduplicateListAuthor = []
    for quote in quotesSoup:
        unduplicateListAuthor.append(quote.find('small',class_="author").text)
    return set(unduplicateListAuthor)

NUMBER_OF_QUOTES=41 
def get_quotes_by_author(author_name, num_quotes=NUMBER_OF_QUOTES):
    base_url = "http://quotes.toscrape.com/"
    quotes = []
    page = 1  # Start from the first page
    
    while len(quotes) < num_quotes:
        # Construct the URL for the page
        url = f"{base_url}page/{page}/"
        response = requests.get(url)
        
        # If the request is successful (status code 200)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # Find all quote elements
            quote_elements = soup.find_all('div', class_='quote')
            
            for quote in quote_elements:
                text = quote.find('span', class_='text').text
                author = quote.find('small', class_='author').text
                
                # Only add quotes from the specified author
                if author_name.lower() in author.lower():
                    quotes.append(text)
                
                # Stop if we've gathered the required number of quotes
                if len(quotes) >= num_quotes:
                    break
        
        # Move to the next page
        page += 1
        
    return quotes

def tacgiaLink(authorName):
    returnAuthor = {
        "TacGia":authorName,
        "Link":"",
        "Namsinh":"",
        "Quote":[]
    }
    for quote in quotes:
        if quote.find('small',class_="author").text == authorName:
            authorLink = f'https://quotes.toscrape.com/{quote.find('a').attrs['href']}'
            
            returnAuthor["Link"] = authorLink
            
            birthResponse = requests.get(authorLink)
            birthSoup = BeautifulSoup(birthResponse.text,'html.parser')
            
            returnAuthor["Namsinh"] = birthSoup.find('span',class_="author-born-date").text
            
            returnAuthor["Quote"] = get_quotes_by_author(authorName)
            break
    return returnAuthor
```

Task 1.2.d) Đưa tất cả thông tin của mỗi tác giả vào `.csv` file

```python
import csv

authorList = unduplicateListAuthor(quotes)

with open('Quote.csv','w') as outputCSV:
    fieldnames = ["TacGia","Link","Namsinh","Quote"]
    writer = csv.DictWriter(outputCSV, fieldnames=fieldnames)
    writer.writeheader()
    for idx in authorList:
        writer.writerow(tacgiaLink(idx))
```
